## 5. Combing the orders_products_combined data with the products data

```python
#Removing the Unnamed column from the products data frame
df_prods.drop(columns = 'Unnamed: 0', inplace=True)
```

```python
#Removing the merged indicator from the dataset
#We want to create a new indicator column for the upcoming merge
df_merged_large.drop(columns = '_merge', inplace=True)
```

```python
df_prods.head()
```

```python
df_merged_large.head()
```

```python
#Left merging the orders dataset onto the clean products dataset
df_ords_prods = df_merged_large.merge(df_prods, on = 'product_id', indicator=True)
```
