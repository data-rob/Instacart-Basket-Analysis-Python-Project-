## Writing code

```python
#Writing price label code

def price_label(row):
    if row['prices'] <= 5:
        return 'Low-range product'
    elif row['prices'] > 5 and row['prices'] <= 15:
        return 'Mid-range product'
    elif row['prices'] > 15:
        return 'High-range product'
    else:
        return 'Not enough data'
```

```python
#Creating new column for the price label

df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts()
```

## Using the .loc() function

```python
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] > 5) & (df['prices'] <= 15), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts()
```

## 2. Creating new busy and slow day column

```python
ords_prods_merged['order_day_of_week'].value_counts()
```

```python
#Adding the two busiest and the two slowest days into a new list

result_new = []

for x in ords_prods_merged['order_day_of_week']:
    if x in (0, 1):
        result_new.append('Busiest days')
    elif x in (3, 4):
        result_new.append('Least busy')
    else:
        result_new.append('Regular busy')
```

```python
ords_prods_merged['busiest day'] = result_new
```

## 4.Identifying busiest hours of the day

```python
#Checking the frequency distribution

ords_prods_merged['order_hour_of_day'].value_counts()
```

```python
#Creating a new list and filling it with strings given conditions

busy_hours = []

for x in ords_prods_merged['order_hour_of_day']:
    if x in (10, 11, 14):
        busy_hours.append('Most orders')
    elif x in (2, 4, 3):
        busy_hours.append('Fewest orders')
    else:
        busy_hours.append('Average orders')
```

```python
#Inputting the strings into a new column in the data frame

ords_prods_merged['busiest_period_of_day'] = busy_hours
```
