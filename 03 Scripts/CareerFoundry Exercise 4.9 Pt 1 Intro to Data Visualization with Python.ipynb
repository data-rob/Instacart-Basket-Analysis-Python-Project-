## 6. Combining the data sets

```python
#Importing the orders and products data set

ords_prods_merged = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_merged_new_columns.pkl'))
```

```python
ords_prods_merged.head()
```

```python
#Checking the user_id column data type to see if it matches the column from the customer data set

ords_prods_merged.info()
```

```python
#Removing the merged column to create a new one

ords_prods_merged.drop(columns = '_merge', inplace=True)
```

```python
#Confirming the column drop

ords_prods_merged.head()
```

```python
#Using a left merge so that all the orders & products data is kept

new_ords_prods_merged = ords_prods_merged.merge(customers, how = 'left', on = 'user_id', indicator = True)
```

```python
new_ords_prods_merged.head()
```

```python
#Checking the merge

new_ords_prods_merged['_merge'].value_counts()
```

All data was merged succesfully. Excellent!
